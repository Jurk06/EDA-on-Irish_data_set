```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from subprocess import check_output
#print(check_output(["ls", "../input"]).decode("utf8"))
```

```
url='https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
df=pd.read_csv(url)
df.head()
```

create a pairplot (plot across different attribute)
size decide the size of the graph 
palette deciede the color

```
a=sns.pairplot(df, hue='species', palette="muted", size=4,  vars=['sepal_length','sepal_width', 'petal_length', 'petal_width'],
               kind='scatter')
a=a.map_offdiag(plt.scatter, s=35, alpha=0.9)
sns.despine()
plt.subplots_adjust(top=0.9)
a.fig.suptitle('Relation btw Sepal width and Sepal length', fontsize=20, color='b', alpha=0.5)
```

```
b=sns.jointplot(x='sepal_width', y='sepal_length', data=df, size=8)
```
