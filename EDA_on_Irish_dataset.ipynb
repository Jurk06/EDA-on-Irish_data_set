```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
df=pd.read_csv(url)
df.head()
```

```python
sns.boxplot(x='species',y='petal_length', data=df)
plt.show()
```

```python
sns.boxplot(x='species', y='petal_width', data=df, linewidth=3.0)
plt.show()
```

```python
sns.violinplot(x='species', y='petal_length', data=df, size=20)
plt.show()
```

```python
sns.swarmplot(x='species', y='petal_length', data=df, color=".35")
plt.show()
```

```python
sns.jointplot(x='petal_length',y='petal_width', data=df,kind='kde')
plt.show()
```

```python
sns.pairplot(data=df)
```
